# Computing the chrombigrams from MIDI
(Skip unless you want to compute it again)

```python
root_dir = 'D:\\MIREX\\all_datasets_raw\\albrecht_symbolic'
annotation_dir = os.path.join(root_dir, "annotations")
file_dir = os.path.join(root_dir, "features")
files_annotations = zip(
    os.listdir(annotation_dir), 
    os.listdir(file_dir)
)
dataset = {
    'files': [],
    'labels': [],
    'chrombigram_sequences': []
}
for annotation_filename, filename in files_annotations:
    print('Parsing ' + filename)    
    dataset['files'].append(filename)
    annotation_path = os.path.join(annotation_dir, annotation_filename)
    filename_path = os.path.join(file_dir, filename)
    with open(annotation_path) as f:
        label = f.read().strip()
    # Parse humdrum file with music21
    pcss = chrombigram.utilities.getpcs_midi(filename_path)    
    chrombis = [Chrombigram(pcs) for pcs in pcss]
#     chrombis_str = ' '.join([str(c) for c in chrombis])
    dataset['labels'].append(label)
    dataset['chrombigram_sequences'].append(chrombis)    
```

```python
# Turn the data into a Pandas DataFrame
df = pd.DataFrame(dataset)
```

```python
# and save it to avoid recomputing
df.to_pickle('dataset.pkl')
```
