## Key Estimation using pitch class sets

This is a small experiment using word embeddings with musical information.

The music pieces are encoded as sequences of `pitch class sets`, which are turned into strings.

This is achieved through a library called `chrombigram` that I developed.

```python
from chrombigram.chrombigram import Chrombigram, ChrombigramCounter
import pandas as pd
import music21
import os
```

## Loading the data

An auxiliary function to extract pitch-class sets from a music21 stream

```python
def getpcs_music21(m21):
    allpcs = []
    for x in m21.chordify().flat.notes:
        pcs = []
        for n in x:
            pcs.append(n.pitch.pitchClass)
        pcs = frozenset(pcs)
        allpcs.append(pcs)
    return allpcs
```

```python
root_dir = 'D:\\MIREX\\all_datasets_raw\\albrecht_kern'
annotation_dir = os.path.join(root_dir, "annotations")
file_dir = os.path.join(root_dir, "features")
files_annotations = zip(
    os.listdir(annotation_dir), 
    os.listdir(file_dir)
)
y = []
X = []
for annotation_filename, filename in files_annotations:
    annotation_path = os.path.join(annotation_dir, annotation_filename)
    filename_path = os.path.join(file_dir, filename)
    with open(annotation_path) as f:
        label = f.read().strip()
    with open(filename_path) as f:
        humdrum = f.read()
    # Parse humdrum file with music21
    m21 = music21.converter.parseData(humdrum, format='humdrum')
    pcs = getpcs_music21(m21)
    chrombis = [Chrombigram(pc) for pc in pcs]
    chrombis_str = ' '.join([str(c) for c in chrombis])
    y.append(label)
    X.append(chrombis_str)
```
